```python
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files

def state_impact_on_accidents(dfs, top_n=5):
    """
    Find and visualize the states with the highest impact on accident numbers.

    Parameters:
    dfs: Dictionary of dataframes from all sheets
    top_n: Number of top states to display (default=5)
    """
    # Create an empty Series to hold state accident counts
    state_accidents = pd.Series(dtype=int)

    # Combine data from all sheets and count accidents by state
    for sheet_name, data in dfs.items():
        if 'State' in data.columns:
            # Add state counts to the series
            state_accidents = state_accidents.add(data['State'].value_counts(), fill_value=0)

    # Sort states by total accident count in descending order
    state_accidents = state_accidents.sort_values(ascending=False)

    # Plotting the top N states with highest accident numbers with increased font sizes
    plt.figure(figsize=(15, 8), dpi=300)
    state_accidents.head(top_n).plot(kind='barh', color='skyblue', edgecolor='black')

    # Title and axis labels with larger font sizes
    plt.title(f'Top {top_n} States with Highest Accident Numbers', fontsize=20)
    plt.xlabel('Total Number of Accidents', fontsize=16)
    plt.ylabel('State', fontsize=16)

    # Increase tick font sizes
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.gca().invert_yaxis()  # Invert y-axis for better readability

    # Annotate total accident numbers on the bars with larger font size
    for index, value in enumerate(state_accidents.head(top_n)):
        plt.text(value + 5, index, f"{int(value)}", fontsize=14, va='center')

    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.tight_layout()

    # Define the filename and save the plot as JPEG with 300 DPI
    file_name = 'Top_5_States_with_Highest_Accident_Numbers.jpeg'
    plt.savefig(file_name, dpi=300, format='jpeg')

    # Display the plot
    plt.show()

    # Automatically download the saved image in Google Colab
    files.download(file_name)

# Run the function for the top 5 states
state_impact_on_accidents(dfs, top_n=5)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

def top_10_separate_visualization(dfs, top_states, temp_data):
    """
    Create two side-by-side horizontal bar charts for accidents and average temperatures for the top 10 states,
    with values annotated on each bar.
    """
    # Initialize a dictionary to store accident counts for top states
    state_accidents = pd.Series(dtype=int)

    # Aggregate accident counts for the top states from 2015 to 2021
    for sheet_name, data in dfs.items():
        if 'State' in data.columns and 'EventDate' in data.columns:
            data['EventDate'] = pd.to_datetime(data['EventDate'], errors='coerce')
            data['Year'] = data['EventDate'].dt.year

            # Filter data for the years 2015-2021 and top states
            year_data = data[(data['Year'] >= 2015) & (data['Year'] <= 2021) & (data['State'].isin(top_states))]
            state_accidents = state_accidents.add(year_data['State'].value_counts(), fill_value=0)

    # Extract average temperature data for the top states
    avg_temp_series = pd.Series({state: temp_data[state]['Avg Temp'] for state in top_states})

    # Plotting the two side-by-side horizontal bar charts
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8), gridspec_kw={'width_ratios': [2, 1]})

    # Plot accident data as a horizontal bar chart with annotations
    ax1.barh(state_accidents.index, state_accidents.values, color='skyblue', edgecolor='black')
    ax1.set_xlabel('Total Number of Accidents', fontsize=14)
    ax1.set_ylabel('State', fontsize=14)
    ax1.set_title('Total Accidents by State (2015-2021)', fontsize=16)
    for index, value in enumerate(state_accidents.values):
        ax1.text(value + 10, index, str(int(value)), va='center', fontsize=12, color='black')

    # Plot average temperature data as a horizontal bar chart with annotations
    ax2.barh(avg_temp_series.index, avg_temp_series.values, color='lightcoral', edgecolor='black')
    ax2.set_xlabel('Average Temperature (°F)', fontsize=14)
    ax2.set_title('Average Temperature by State (2015-2021)', fontsize=16)
    ax2.set_yticks([])  # Hide y-axis labels on the second chart to avoid redundancy
    for index, value in enumerate(avg_temp_series.values):
        ax2.text(value + 0.5, index, f"{value:.1f}", va='center', fontsize=12, color='black')

    plt.tight_layout()

    # Save and download the figure as JPEG with 300 dpi
    file_name = "top_10_states_accidents_and_temp_2015_2021_annotated.jpg"
    plt.savefig(file_name, dpi=300, format="jpeg", bbox_inches="tight")
    plt.show()
    files.download(file_name)

# Run the function with the loaded data
top_10_separate_visualization(dfs, top_states, temp_data_example)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

def top_10_combined_visualization_with_annotations(dfs, top_states, temp_data):
    """
    Create a single horizontal bar chart for accidents, with annotations for both accident counts and temperatures.
    """
    # Initialize a dictionary to store accident counts for top states
    state_accidents = pd.Series(dtype=int)

    # Aggregate accident counts for the top states from 2015 to 2021
    for sheet_name, data in dfs.items():
        if 'State' in data.columns and 'EventDate' in data.columns:
            data['EventDate'] = pd.to_datetime(data['EventDate'], errors='coerce')
            data['Year'] = data['EventDate'].dt.year

            # Filter data for the years 2015-2021 and top states
            year_data = data[(data['Year'] >= 2015) & (data['Year'] <= 2021) & (data['State'].isin(top_states))]
            state_accidents = state_accidents.add(year_data['State'].value_counts(), fill_value=0)

    # Extract average temperature data for the top states
    avg_temp_series = pd.Series({state: temp_data[state]['Avg Temp'] for state in top_states})

    # Sort the data to ensure consistency
    state_accidents = state_accidents.sort_index()
    avg_temp_series = avg_temp_series.sort_index()

    # Plotting the single horizontal bar chart with annotations
    fig, ax = plt.subplots(figsize=(14, 8))

    # Plot accident data as a horizontal bar chart
    bars = ax.barh(state_accidents.index, state_accidents.values, color='skyblue', edgecolor='black')
    ax.set_xlabel('Total Number of Accidents', fontsize=14)
    ax.set_ylabel('State', fontsize=14)
    ax.set_title('Total Accidents and Average Temperatures by State (2015-2021)', fontsize=16)

    # Annotate each bar with accident count and temperature
    for index, (accident, temp) in enumerate(zip(state_accidents.values, avg_temp_series.values)):
        # Position accident count above the bar
        ax.text(accident + 10, index, f"Accidents: {int(accident)}", va='center', fontsize=12, color='black')
        # Position temperature below the bar
        ax.text(accident + 10, index - 0.25, f"Temp: {temp:.1f}°F", va='center', fontsize=12, color='red')

    plt.tight_layout()

    # Save and download the figure as JPEG with 300 dpi
    file_name = "top_10_states_accidents_and_temp_combined_2015_2021.jpg"
    plt.savefig(file_name, dpi=300, format="jpeg", bbox_inches="tight")
    plt.show()
    files.download(file_name)

# Run the function with the loaded data
top_10_combined_visualization_with_annotations(dfs, top_states, temp_data_example)
```
