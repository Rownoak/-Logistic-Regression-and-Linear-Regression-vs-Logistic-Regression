```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import time

# Define the 3-layer model architecture with ReLU as the activation function
model = Sequential([
    Dense(256, input_dim=X_train.shape[1], activation='relu'),  # First hidden layer with 256 neurons
    Dense(2, activation='relu'),                               # Second hidden layer with 2 neurons
    Dense(1, activation='linear')                              # Output layer for regression
])

# Specify the optimizer (Adam) and loss (Mean Squared Error)
optimizer = Adam(learning_rate=0.001)  # Learning rate can be adjusted as needed
model.compile(optimizer=optimizer, loss='mean_squared_error')

# Print model summary to understand the architecture
model.summary()

# Measure the time to check for convergence
start_time = time.time()

# Fit the model until convergence (100 epochs, batch size = 128)
history = model.fit(X_train, y_train, epochs=100, batch_size=128, validation_data=(X_val, y_val), verbose=1)

# Measure the time taken
end_time = time.time()
print(f"Training completed in {end_time - start_time:.2f} seconds")

# Predict on the test data
y_pred = model.predict(X_test)

# Calculate the Test MSE
test_mse = mean_squared_error(y_test, y_pred)
print(f"Test MSE: {test_mse}")

# Plot training and validation loss curves
plt.plot(history.history['loss'], label='Training Loss', color='blue')
plt.plot(history.history['val_loss'], label='Validation Loss', color='orange')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error Loss')
plt.legend()
plt.grid(True)
plt.title('Training and Validation Loss over Epochs')

# Get the final loss values
final_training_loss = history.history['loss'][-1]
final_validation_loss = history.history['val_loss'][-1]

# Annotate the final loss values on the plot
plt.annotate(f'Final Training Loss: {final_training_loss:.4f}',
             xy=(len(history.history['loss'])-1, final_training_loss),
             xytext=(len(history.history['loss'])-15, final_training_loss+0.5),
             arrowprops=dict(facecolor='blue', shrink=0.05), fontsize=12, color='blue')

plt.annotate(f'Final Val Loss: {final_validation_loss:.4f}',
             xy=(len(history.history['val_loss'])-1, final_validation_loss),
             xytext=(len(history.history['val_loss'])-15, final_validation_loss+0.5),
             arrowprops=dict(facecolor='orange', shrink=0.05), fontsize=12, color='orange')

# Show the plot
plt.show()
```

```python
import time  # To calculate computation time

# Function to train and evaluate the model with different batch sizes
def evaluate_batch_size(batch_size):
    print(f"\nEvaluating with batch size: {batch_size}")

    # Create the 3-layer model architecture (256-2-1 neurons)
    model = Sequential([
        Dense(256, input_dim=X_train.shape[1], activation='relu'),  # First hidden layer with 256 neurons
        Dense(2, activation='relu'),                               # Second hidden layer with 2 neurons
        Dense(1, activation='linear')                              # Output layer for regression
    ])

    # Compile the model with Adam optimizer and MSE loss function
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Measure start time for computation
    start_time = time.time()

    # Train the model with the given batch size
    history = model.fit(X_train, y_train, epochs=100, batch_size=batch_size, validation_data=(X_val, y_val), verbose=0)

    # Measure end time and calculate computation time
    end_time = time.time()
    computation_time = end_time - start_time
    print(f"Computation time for batch size {batch_size}: {computation_time:.2f} seconds")

    # Predict on the test data
    y_pred = model.predict(X_test)

    # Calculate Test MSE
    test_mse = mean_squared_error(y_test, y_pred)
    print(f"Test MSE for batch size {batch_size}: {test_mse}")

    return test_mse, computation_time

# Evaluate with batch sizes {32, 64, 128, 256}
batch_sizes = [32, 64, 128, 256]
results = {}

for batch_size in batch_sizes:
    test_mse, computation_time = evaluate_batch_size(batch_size)
    results[batch_size] = {"Test MSE": test_mse, "Computation Time (s)": computation_time}

# Print the results for all batch sizes
print("\nResults for different batch sizes:")
for batch_size, metrics in results.items():
    print(f"Batch Size: {batch_size} - Test MSE: {metrics['Test MSE']:.4f}, Computation Time: {metrics['Computation Time (s)']:.2f} seconds")
```

```python
# Function to train and evaluate the model with different learning rates (α)
def evaluate_learning_rate(alpha):
    print(f"\nEvaluating with learning rate (α): {alpha}")

    # Create the 3-layer model architecture (256-2-1 neurons)
    model = Sequential([
        Dense(256, input_dim=X_train.shape[1], activation='relu'),  # First hidden layer with 256 neurons
        Dense(2, activation='relu'),                               # Second hidden layer with 2 neurons
        Dense(1, activation='linear')                              # Output layer for regression
    ])

    # Compile the model with Adam optimizer and varying learning rate
    optimizer = tf.keras.optimizers.Adam(learning_rate=alpha)
    model.compile(optimizer=optimizer, loss='mean_squared_error')

    # Measure start time for computation
    start_time = time.time()

    # Train the model with the given learning rate
    history = model.fit(X_train, y_train, epochs=100, batch_size=128, validation_data=(X_val, y_val), verbose=0)

    # Measure end time and calculate computation time
    end_time = time.time()
    computation_time = end_time - start_time
    print(f"Computation time for learning rate (α) {alpha}: {computation_time:.2f} seconds")

    # Predict on the test data
    y_pred = model.predict(X_test)

    # Calculate Test MSE
    test_mse = mean_squared_error(y_test, y_pred)
    print(f"Test MSE for learning rate (α) {alpha}: {test_mse}")

    return test_mse, computation_time

# Evaluate with learning rates {0.1, 0.01, 0.0001}
learning_rates = [0.1, 0.01, 0.0001]
results_lr = {}

for alpha in learning_rates:
    test_mse, computation_time = evaluate_learning_rate(alpha)
    results_lr[alpha] = {"Test MSE": test_mse, "Computation Time (s)": computation_time}

# Print the results for all learning rates
print("\nResults for different learning rates:")
for alpha, metrics in results_lr.items():
    print(f"Learning Rate (α): {alpha} - Test MSE: {metrics['Test MSE']:.4f}, Computation Time: {metrics['Computation Time (s)']:.2f} seconds")
```

```python
from tensorflow.keras.layers import Dropout
import time  # For measuring computation time

# Function to train and evaluate the model with different dropout rates
def evaluate_dropout(dropout_rate):
    print(f"\nEvaluating with Dropout rate: {dropout_rate}")

    # Create the 3-layer model architecture with Dropout
    model = Sequential([
        Dense(256, input_dim=X_train.shape[1], activation='relu'),  # First hidden layer with 256 neurons
        Dropout(dropout_rate),                                      # Dropout layer after the first hidden layer
        Dense(2, activation='relu'),                                # Second hidden layer with 2 neurons
        Dropout(dropout_rate),                                      # Dropout layer after the second hidden layer
        Dense(1, activation='linear')                               # Output layer for regression
    ])

    # Compile the model with Adam optimizer and MSE loss function
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Measure start time for computation
    start_time = time.time()

    # Train the model with the given dropout rate
    history = model.fit(X_train, y_train, epochs=100, batch_size=128, validation_data=(X_val, y_val), verbose=0)

    # Measure end time and calculate computation time
    end_time = time.time()
    computation_time = end_time - start_time
    print(f"Computation time for Dropout rate {dropout_rate}: {computation_time:.2f} seconds")

    # Predict on the test data
    y_pred = model.predict(X_test)

    # Calculate Test MSE
    test_mse = mean_squared_error(y_test, y_pred)
    print(f"Test MSE for Dropout rate {dropout_rate}: {test_mse}")

    return test_mse, computation_time

# Evaluate with dropout rates {0.2, 0.5, 0.7}
dropout_rates = [0.2, 0.5, 0.7]
results_dropout = {}

for rate in dropout_rates:
    test_mse, computation_time = evaluate_dropout(rate)
    results_dropout[rate] = {"Test MSE": test_mse, "Computation Time (s)": computation_time}

# Print the results for all dropout rates
print("\nResults for different Dropout rates:")
for rate, metrics in results_dropout.items():
    print(f"Dropout Rate: {rate} - Test MSE: {metrics['Test MSE']:.4f}, Computation Time: {metrics['Computation Time (s)']:.2f} seconds")
```
